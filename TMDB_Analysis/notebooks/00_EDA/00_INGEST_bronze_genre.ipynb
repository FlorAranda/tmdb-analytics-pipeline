```sql
%python
import requests

api_key = "e9e16e07083875eb883ce4f6b526a17d"
url = "https://api.themoviedb.org/3/genre/movie/list"

params = {"api_key": api_key}

response = requests.get(url, params=params)

genres = response.json()["genres"]

from pyspark.sql.functions import col, current_timestamp

df_genres = spark.createDataFrame(genres)

df_genres = df_genres.select(
    col("id").alias("genre_id"),
    col("name").alias("genre_name"),
    current_timestamp().alias("ingestion_timestamp")
)

# guardado en volumen (DATA LAKE)
df_genres.write \
    .mode("overwrite") \
    .json("/Volumes/analisis_tmdb/bronze/archivos/genres/")
```
