```sql
INSERT OVERWRITE analisis_tmdb.silver.movie_genres

WITH exploded AS (
    SELECT
        id as movie_id,
        EXPLODE(genre_ids) AS genre_id
    FROM analisis_tmdb.bronze.movies_bronze
),

cleaned AS (
    SELECT
        movie_id,
        genre_id AS genre_id
    FROM exploded
),

deduplicated AS (
    SELECT DISTINCT
        movie_id,
        genre_id
    FROM cleaned
)

SELECT *
FROM deduplicated
order by movie_id;
```
