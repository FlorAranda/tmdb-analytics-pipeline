```sql
INSERT OVERWRITE analisis_tmdb.gold.dim_date

SELECT DISTINCT
    CAST(DATE_FORMAT(release_date, 'yyyyMMdd') AS INT) AS date_key,
    release_date AS full_date,
    YEAR(release_date) AS year,
    QUARTER(release_date) AS quarter,
    MONTH(release_date) AS month,
    DATE_FORMAT(release_date, 'MMMM') AS month_name,
    DAY(release_date) AS day,
    DAYOFWEEK(release_date) AS day_of_week,
    DATE_FORMAT(release_date, 'EEEE') AS day_name
FROM analisis_tmdb.silver.movies
WHERE release_date IS NOT NULL;
```
