```sql
MERGE INTO analisis_tmdb.gold.dim_movies t
USING analisis_tmdb.silver.movies s
ON t.movie_id = s.movie_id

WHEN MATCHED 
AND t.ingestion_timestamp < s.ingestion_timestamp THEN UPDATE SET
    title = s.title,
    original_title = s.original_title,
    overview = s.overview,
    release_date = s.release_date,
    release_year = s.release_year,
    original_language = s.original_language,
    ingestion_timestamp = s.ingestion_timestamp

WHEN NOT MATCHED THEN INSERT (
    movie_id,
    title,
    original_title,
    overview,
    release_date,
    release_year,
    original_language,
    ingestion_timestamp
)
VALUES (
    s.movie_id,
    s.title,
    s.original_title,
    s.overview,
    s.release_date,
    s.release_year,
    s.original_language,
    s.ingestion_timestamp
);
```
