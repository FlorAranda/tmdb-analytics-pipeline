```sql
MERGE INTO analisis_tmdb.gold.dim_genres t
USING analisis_tmdb.silver.genres s
ON t.genre_id = s.genre_id

WHEN MATCHED 
AND t.ingestion_timestamp < s.ingestion_timestamp THEN UPDATE SET
    genre_name = s.genre_name,
    ingestion_timestamp = s.ingestion_timestamp

WHEN NOT MATCHED THEN INSERT (
    genre_id,
    genre_name,
    ingestion_timestamp
)
VALUES (
    s.genre_id,
    s.genre_name,
    s.ingestion_timestamp
);
```
