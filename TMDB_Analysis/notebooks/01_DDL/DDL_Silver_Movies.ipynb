```sql
%sql

-- ============================================================
-- CREAR TABLA SILVER: movies
-- ============================================================

DROP TABLE IF EXISTS analisis_tmdb.silver.movies;

CREATE TABLE analisis_tmdb.silver.movies (

    movie_id BIGINT COMMENT 'ID único de la película',
    title STRING COMMENT 'Título localizado de la película',
    original_title STRING COMMENT 'Título original de la película',
    overview STRING COMMENT 'Descripción o sinopsis de la película',
    release_date DATE COMMENT 'Fecha de estreno',
    release_year INT COMMENT 'Año de estreno',
    original_language STRING COMMENT 'Idioma original',
    popularity DECIMAL(10,2) COMMENT 'Popularidad según TMDB',
    vote_average DECIMAL(3,2) COMMENT 'Promedio de votos',
    vote_count INT COMMENT 'Cantidad total de votos',
    ingestion_timestamp TIMESTAMP COMMENT 'Timestamp de ingesta'

)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla silver con datos limpios de películas';
```
