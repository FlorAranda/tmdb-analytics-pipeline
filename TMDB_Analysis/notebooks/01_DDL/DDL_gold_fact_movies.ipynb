```sql
CREATE OR REPLACE TABLE analisis_tmdb.gold.fact_movies (
    movie_key BIGINT,   -- FK → dim_movies
    genre_key BIGINT,   -- FK → dim_genres
    date_key INT,       -- FK → dim_date
    popularity DECIMAL(10,2),
    vote_average DECIMAL(3,2),
    vote_count INT,
    rating_weighted DECIMAL(10,2),
    ingestion_timestamp TIMESTAMP
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla FACT de peliculas';
```
