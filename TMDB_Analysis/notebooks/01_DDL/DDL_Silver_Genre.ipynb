```sql
-- ============================================================
-- CREAR TABLA SILVER: genre
-- ============================================================

DROP TABLE IF EXISTS analisis_tmdb.silver.genres;

CREATE TABLE analisis_tmdb.silver.genres (
    genre_id INT COMMENT 'ID del género',
    genre_name STRING COMMENT 'Nombre del género',
    ingestion_timestamp TIMESTAMP COMMENT 'Timestamp de ingesta'
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla de dimensión de géneros';
```
