```sql
-- ============================================================
-- Creacion de catálogo
-- ============================================================

CREATE CATALOG IF NOT EXISTS analisis_tmdb;
```

```sql
-- ============================================================
-- Creacion del esquema BRONZE
-- ============================================================

USE CATALOG analisis_tmdb;

-- Crear esquema bronze
CREATE SCHEMA IF NOT EXISTS analisis_tmdb.bronze
COMMENT 'Bronze layer - datos crudos provenientes de la API';

-- Verificar
SHOW SCHEMAS;
```

```sql
-- ============================================================
-- Creacion de esquema Silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS analisis_tmdb.silver
COMMENT 'Capa Silver: Datos limpios, validados y con tipos correctos';

-- Verificar creación
DESCRIBE SCHEMA EXTENDED analisis_tmdb.silver;
```

```sql
-- ============================================================
-- Creacion de esquema Gold
-- ============================================================

CREATE SCHEMA IF NOT EXISTS analisis_tmdb.gold
COMMENT 'Capa Gold: Datos agregados y métricas para negocio';

-- Verificar creación
DESCRIBE SCHEMA EXTENDED analisis_tmdb.gold;
```

```sql
-- ============================================================
-- Creacion de Volume para archivos
-- ============================================================

CREATE VOLUME IF NOT EXISTS analisis_tmdb.bronze.archivos
COMMENT 'Volume para almacenar archivos crudos';

-- Verifico que se creó
SHOW VOLUMES IN analisis_tmdb.bronze;
```
