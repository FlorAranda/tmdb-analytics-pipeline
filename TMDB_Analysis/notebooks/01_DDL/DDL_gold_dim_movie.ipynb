```sql
CREATE OR REPLACE TABLE analisis_tmdb.gold.dim_movies (
    movie_key BIGINT GENERATED ALWAYS AS IDENTITY,
    movie_id INT,
    title STRING,
    original_title STRING,
    overview STRING,
    release_date DATE,
    release_year INT,
    original_language STRING,
    ingestion_timestamp TIMESTAMP
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla DIM de peliculas';
```
