```sql
-- ============================================================
-- CREAR TABLA SILVER: movies_genres
-- ============================================================

DROP TABLE IF EXISTS analisis_tmdb.silver.movie_genres;

CREATE TABLE analisis_tmdb.silver.movie_genres (
    movie_id BIGINT COMMENT 'ID de la película',
    genre_id INT COMMENT 'ID del género'
)
USING DELTA
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported')
COMMENT 'Tabla puente entre películas y géneros';
```
